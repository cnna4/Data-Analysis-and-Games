```python
import pandas as pd
import numpy as np
import seaborn as sb
import matplotlib.pyplot as plt
#Importing needed libraries
```

```python
#We read the csv file and split the data by state.
data = pd.read_csv('United_States_COVID-19_Cases_and_Deaths_by_State_over_Time.csv')
TX = data.state.str.contains('TX')
NY = data.state.str.contains('NY')
CT = data.state.str.contains('CT')
RI = data.state.str.contains('RI')
CA = data.state.str.contains('CA')
FL = data.state.str.contains('FL')
GA = data.state.str.contains('GA')
TX = data[TX]
NY = data[NY]
CT = data[CT]
RI = data[RI]
CA = data[CA]
FL = data[FL]
GA = data[GA]
#Adjusting fon size to read the y axis.
font = {'family' : 'DejaVu Sans',
        'weight' : 'bold',
        'size'   : 35}

matplotlib.rc('font', **font)
```

```python
plt.plot(TX.submission_date, TX.new_case)
plt.show()


#Texas Reopened May 1, then scaled back on July 3 after a resurgence. Worst day was Sept. 22, one day after increasing
#business operations again
```

```python
plt.plot(FL.submission_date, FL.new_case)
plt.show()

#Reopened May 4, then went one step further on June 5
#Worst stretch between late June - July
#Phase three started Sept. 25, lifted most restrictions
```

```python
plt.plot(GA.submission_date, GA.new_case)
plt.show()

#One of the first to reopen on April 24
#Opposition to mask mandate
#Back and forth between state and Atlanta
```

```python
plt.plot(CA.submission_date, CA.new_case)
plt.show()

#Reopened May 25, backtracked after surge in cases
#Now classifying counties to see if they can reopen or not
```

```python
plt.plot(NY.submission_date, NY.new_case)
plt.show()

#One of the few states to reopen after cases went down
#Did not allow indoor dining until last week
#Stagnant in case since the reopening
```

```python
plt.plot(CT.submission_date, CT.new_case)
plt.show()

#Reopened May 20
#Delayed phase 3 in June due to rise in cases in other states, will be in effect October 8
#Strict list of blacklisted states 
```

```python
plt.plot(RI.submission_date, RI.new_case)
plt.show()

#Reopened May 9
#Reached phase 3 of its reopening rather quickly on June 30
```
